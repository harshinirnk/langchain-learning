# **LLM CALL**

```python
#LLM call using Langchain
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv


llm_call=ChatOpenAI(
    model="gpt-4.1",
    api_key=os.getenv("tokens"),
    base_url="https://models.inference.ai.azure.com",
    temperature=0
)
response=llm_call.invoke("write code for adding two numbers in python")
print(response.content)
```
